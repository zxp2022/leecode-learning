给你单链表的头节点 head ，请你反转链表，并返回反转后的链表。

思路一:既然链表不容易反向操作,那就把它变成列表,反转列表之后重新创建一个新的链表

````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        def get_len(node):
            count = 0
            p = node 
            while p:
                count += 1
                p = p.next
            return count
        list_len = get_len(head)
        visited = [] * list_len
        p = head
        while p:
            visited.append(p)
            p = p.next
        dummy = ListNode(0)
        curr = dummy
        for i in visited[::-1]:
            curr.next = i
        return dummy.next
````

``dummy`` 哑节点: 是一个假的,不存真实数据的节点,存在是为了消除“头节点”的特殊性，让所有节点都能被一视同仁地处理.

有很多错误:
1. `append()` 会自动扩容,不需要获取长度
2. 指针原地踏步
```python
for i in visited[::-1]:
    curr.next = i
    # ⚠️ 缺了一行代码！curr 从来没有移动过！
```
3. 存进 visited 列表的是 原来的节点对象  
原来的链表：1 -> 2  
你的反转操作：curr.next = i，你让 2 指向了 1  
形成了 1 <-> 2 的环  


````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        #缺少if not head: 处理输入一开始就是 [] 的情况，防止 head.next 报错

        visited = []
        p = head
        while p:
            visited.append(p)
            p = p.next
        dummy = ListNode(0)
        curr = dummy
        for i in visited[::-1]:
            curr.next = i
            curr = curr.next
        head.next = None # 更新旧头节点的指向，否则它还指着原来的第2个节点，形成环。
        return dummy.next
````

会报错,因为试图在 `None` 身上找 `next`属性


````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head: return None
        visited = []
        p = head
        while p:
            visited.append(p)
            p = p.next
        dummy = ListNode(0)
        curr = dummy
        for i in visited[::-1]:
            curr.next = i
            curr = curr.next
        
        head.next = None
        
        return dummy.next
````

列表里放的是节点的引用,可以放节点的值就不用切断连接了.

````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head: return None

        # 把链表变成列表
        p = head
        visited = []
        while p:
            visited.append(p.value)
            p = p.next

        #反转列表
        visited.reverse()
        
        #重新构建链表
        dummy = ListNode(0)
        curr = dummy
        for i in visited:
            curr.next = ListNode(i)
            curr = curr.next
        return dummy.next        
````

空间浪费严重 $O(N)$,需要单独存一个列表.可优化至 $O(1)$

双指针迭代法: 在遍历链表的时候,把每个节点的next掰过来,指向它的前一个节点

需要三个角色:
1. `prev` 前一个节点,一开始是None
2. `curr` 当前节点,一开始是head
3. `temp` 临时存下一个节点,防止断链后找不到路

````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None # 前驱节点,初始为空
        curr = head # 当前节点,初始为头结点
        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp
        #循环结束时curr变成None,prev刚好停在头结点上
        return prev
````

只用了两三个变量,没有额外申请内存
