给你一个 m 行 n 列的矩阵 matrix ，请按照 顺时针螺旋顺序 ，返回矩阵中的所有元素。

有两种解法：设定边界法和削皮旋转法
其中最推荐掌握的是设定边界法，是解决所有矩阵遍历问题的通解。


````python
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        if not matrix:
            return []
        ans = []
        #1.设定4个边界
        top, bottom = 0, len(matrix) - 1
        left, right = 0, len(matrix[0]) - 1
        while True:
            #向右走
            for i in range(left, right + 1):
                ans += matrix[top][i]
            top += 1
            if top > bottom:
                break
            #向下走
            for i in range(top, bottom + 1):
                ans += matrix[i][right]
            right -= 1
            if right < left: break
            #向左走,切记range左闭右开
            for i in range(right, left - 1, -1): 
                ans += matrix[top][i]
            bottom -= 1
            if bottom < top: break
            #向上走
            for i in range(bottom, top - 1, -1):
                ans += matrix[i][left]
            left += 1
            if left > right: break
        return ans
````

上述解法犯了两个错误：
1. `+=` 的意思是：把右边那个箱子里的东西全倒进左边的箱子  
    我企图用 `list(matrix[][])` 来把数字打包进一个列表里，这是不对的。`list() ` 要做的是把里面的东西拆开，然后把拆出来的零件一个个放进新的列表里
    可以 `[matrix[][]]` ,也可以用 `.append() `
2. 向左走时是逻辑错误
    ```diff
    - ans += matrix[top][i]
    + ans += matrix[bottom][i]
    ```

边界法需要练得滚瓜烂熟。它是解决所有矩阵遍历问题的通解！  
经历四个步骤：循环遍历，添加元素，缩墙，检查死路  
空间复杂度：$O(1)$


旋转削皮法：不断取走第一行`matrix.pop(0)`，再将剩下的矩阵逆时针旋转 90 度，循环往复直到矩阵为空   
矩阵逆时针旋转：`list(zip(*matrix))[::-1]`

````python
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        if not matrix:
            return []
        ans = []
        m, n = len(matrix), len(matrix[0])
        for i in range(m + n - 1):
            ans += matrix.pop(0)
            matrix = zip(*matrix)[::-1] # 矩阵逆时针旋转
        return ans
````

犯了两个错误：
1. zip对象不可切片，`zip()` 返回的是一个迭代器不是列表不知道长度,所以不支持切片操作
    可以用 `list(zip(*matrix))[::-1]` 解决
2. 循环次数不对，应该改为 `while matrix:`
