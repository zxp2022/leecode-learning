给定一个 m x n 的矩阵，如果一个元素为 0 ，则将其所在行和列的所有元素都设为 0 。请使用 原地 算法。

````python
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        col_flags, row_flags = [], []
        for col in zip(*matrix):
            if 0 in col:
                col_flags += [1]
            else:
                col_flags += [0]
        for row in matrix:
            if 0 in row:
                row_flags += [1]
            else:
                row_flags += [0]
        for i, r in enumerate(row_flags):
            for j, c in enumerate(col_flags):
                if c or r:
                    matrix[i][j] = 0      
````

空间复杂度是 $O(M+N)$ 需要优化到 $O(1)$

````python
class Solution:
    def setZeros(self, matrix):
        m, n = len(matrix), len(matrix[0])
        col0_flags = 1 #第一列是否有零单独存储
        for i in range(m):
            for j in range(1, n):
                if matrix[i][0] == 0:
                    col0_flags = 0
                if matrix[i][j] == 0:
                    matrix[0][j], matrix[i][0] = 0, 0
                    #此时已经获得矩阵第一行储存(除第一列)每一列是否有零
                    #第一列储存每一行是否有零
        for i in range(m - 1, 0, -1):
            for j in range(n-1, 0, -1):
                if not matrix[i][0] or not matrix[0][j]:
                    matrix[i][j] = 0
        for i in range(1, m):
            if col0_flags == 0:
                matrix[i][0] =0 
        for j in range(1, n):
            if matrix[0][0] == 0:
                matrix[0][j] = 0
````

有很多问题:
1. 第一列的检查位置错误,如果矩阵只有一列,内层循环一次都不会执行.必须放到内层循环j的外面
2. 左上角 `matrix[0][0]`的更新不完整
    ```diff
    - for i in range(1, m):
    + for i in range(0, m):
    ```
3. 逻辑混乱,分为四个for循环块,标准解法将步骤3和4融合进步骤2  
    标记（2层循环）。  
    倒序处理内部（2层循环）。  
    单独处理第一列（1层循环）。  
    单独处理第一行（1层循环）。

````python
class Solution:
    def setZeros(self, matrix):
        m, n = len(matrix), len(matrix[0])
        col0_flags = 1 #第一列是否有零单独存储
        for i in range(m):
                if matrix[i][0] == 0:
                    col0_flags = 0
            for j in range(1, n):
                if matrix[i][j] == 0:
                    matrix[0][j], matrix[i][0] = 0, 0
                    #此时已经获得矩阵第一行储存(除第一列)每一列是否有零
                    #第一列储存每一行是否有零
        for i in range(m - 1, -1, -1):
            for j in range(n-1, 0, -1):
                if not matrix[i][0] or not matrix[0][j]:
                    matrix[i][j] = 0
            #在处理每一行时顺手处理改行的第一列元素
            if col0_flags == 0:
                matrix[i][0] =0 
````
