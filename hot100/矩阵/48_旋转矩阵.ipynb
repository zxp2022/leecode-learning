给定一个 n × n 的二维矩阵 matrix 表示一个图像。请你将图像顺时针旋转 90 度。

```python
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        matrix[:] = list(zip(*matrix[::-1]))
```

``matrix = zip(*matrix[::-1])`` 表示的一次是把matrix这个门牌号贴到新房间上,就旧房间(要修改的矩阵)纹丝未动  

``[:]`` 表示操作在这个对象内部,它表示“替换 matrix 内部的内容”，而不是让 matrix 变量指向别处

``zip()`` 的功能：把多个列表（或可迭代对象）中对应的元素，“打包”成一个个元组。

``*`` 操作符在这里的作用是 “把容器打散”。

1. ``*matrix`` 把外层的列表拆了，把里面的三行当作 3 个独立的参数 传给 ``zip``。
2. ``zip`` 拿到了 3 个列表，开始工作：把三个列表中的对应元素打包成一个个元组

``matrix[::-1]`` 实现矩阵上下翻转

python不支持类似``matrix[:][::-1]`` 的多维切片

要想实现矩阵水平翻转:``matrix[:] = [row[::-1] for row in matrix]``





```python
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        matrix[:] = zip(*matrix[::-1])
```

```python
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        # 1. matrix[::-1] : 先上下翻转
        # 2. zip(*) : 再转置
        # 3. matrix[:] = : 必须用切片赋值来实现“原地修改” 保留内存地址，把数据元素换成新的
        # .reverse() 方法是 原地修改，它的返回值永远是 None
        matrix[:] = [row[::-1] for row in list(zip(*matrix))]
```
