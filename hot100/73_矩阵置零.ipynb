给定一个 m x n 的矩阵，如果一个元素为 0 ，则将其所在行和列的所有元素都设为 0 。请使用 原地 算法。

```python
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        col_flags, row_flags = [], []
        for col in zip(*matrix):
            if 0 in col:
                col_flags += [1]
            else:
                col_flags += [0]
        for row in matrix:
            if 0 in row:
                row_flags += [1]
            else:
                row_flags += [0]
        for i, r in enumerate(row_flags):
            for j, c in enumerate(col_flags):
                if c or r:
                    matrix[i][j] = 0      
```

有一些问题需要注意:
1. `+=` 性能不如 `append()`
2. 用列表推导式简化

```python
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        col_flags = [0 in col for col in zip(*matrix) ]
        row_flags = [0 in row for row in matrix]
        for i, r in enumerate(row_flags):
            for j, c in enumerate(col_flags):
                if c or r:
                    matrix[i][j] = 0      
```
